```python
# 2013년 1월 9일 14시 채우기
# 해당 월, 일, 시간의 평균 구하기
nan_temp = total.loc[(total['month']==1) & (total['day']==9) & (total['시간']==14)]['기온'].mean()
# 2013년 1월 9일 14시 채우기
total.loc[(total['month']==1) & (total['day']==9) & (total['시간']==14) & (total['기온'].isnull())] = nan_temp
```

```python
# 확인
total.loc[(total['month']==1) & (total['day']==9) & (total['시간']==14) & (total['기온'].isnull())]
```

```python
# 결측치 채우기 함수화
def nan_fill(month, day, time):
    # 해당 월, 일, 시간의 평균 구하기
    nan_temp = total.loc[(total['month']==month) & (total['day']==day) & (total['시간']==time)]['기온'].mean()
    # 2013년 1월 9일 14시 채우기
    total.loc[(total['month']==month) & (total['day']==day) & (total['시간']==time) & (total['기온'].isnull())] = nan_temp
    return total
```

```python
# 확인
total[total['기온'].isnull()][:20]
```

```python
for m in [1,3,5,7,8,10,12]:
    for d in range(1,32):
        for h in range(1,25):
            total = nan_fill(m, d, h)
```
