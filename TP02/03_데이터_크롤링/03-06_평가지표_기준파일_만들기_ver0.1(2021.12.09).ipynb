```python
import pandas as pd
from tqdm import tqdm
```

# 공급량과 외부데이터 자료 가져오기

```python
df = pd.read_csv('../04_가스_기온_합치기/2013-2018년_가스공급량_기온_습도_기압02.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.sort_values(['year','month','day','구분','시간'])
df.head()
```

# 월, 일, 구분, 시간 별 평균 공급량

```python
mean_df = df.groupby(['month', 'day', '구분', '시간'], as_index=False).mean()
mean_df
```

```python
mean_df = mean_df.reset_index(drop=True)
mean_df.info()
```

# 제출 파일 불러오기

```python
sub = pd.read_csv('../data/sample_submission.csv')
sub.info()
```

```python
sub.head()
```

```python
sub[['year', 'month', 'day', '시간', '구분']]= sub['일자|시간|구분'].str.split(r'-| ').tolist()
sub.head()
```

```python
# 연, 월, 일, 시간 숫자형으로 변환
sub['year'] = sub['year'].astype(int)
sub['month'] = sub['month'].astype(int)
sub['day'] = sub['day'].astype(int)
sub['시간'] = sub['시간'].astype(int)
sub.info()
```

```python
sub['구분'].unique()
```

```python
# 구분 숫자로 바꾸기
d_map = {}
for i, d in enumerate(sub['구분'].unique()):
    d_map[d] = i
sub['구분'] = sub['구분'].map(d_map)
```

```python
d_map
```

```python
sub.head()
```

```python
mean_basic = pd.merge(sub, mean_df, how='left', on=['month', 'day', '시간', '구분'])
mean_basic.head()
```

```python
mean_basic.info()
```

```python
mean_basic = mean_basic[['일자|시간|구분','공급량_y']]
mean_basic.columns = [['일자|시간|구분','공급량']]
mean_basic.head()
```

```python
mean_basic.tail()
```

```python
sub.head()
```

# 파일로 만들기

```python
mean_basic.to_csv('mean_basic_sub01.csv', index=False)
```
