### 1. 새로 일자 만들어 데이터셋에 입력하기

```python
date_list2013 = ["2012-12-24 24"]
for m in [1,3,5,7,8,10,12]:
    for d in range(1,32):
        for h in range(1,25):
            date_list2013.append(f"2013-{m:0>2}-{d:0>2} {h:0>2}")

for m in [4,6,9,11]:
    for d in range(1,31):
        for h in range(1,25):
            date_list2013.append(f"2013-{m:0>2}-{d:0>2} {h:0>2}")

for m in [2]:
    for d in range(1,29):
        for h in range(1,25):
            date_list2013.append(f"2013-{m:0>2}-{d:0>2} {h:0>2}")

date_list2013.sort()
date_list2013[:10]
```

```python
date_list2013[-10:]
```

```python
len(date_list2013)
```

### 2013, 2014, 2015, 2017, 2018년 날짜 리스트 만들기
* 2016년은 윤년

```python
date_list = []
for y in [13, 14, 15, 17, 18]:
    for m in [1,3,5,7,8,10,12]:
        for d in range(1,32):
            for h in range(1,25):
                date_list.append(f"20{y:0>2}-{m:0>2}-{d:0>2} {h:0>2}")

    for m in [4,6,9,11]:
        for d in range(1,31):
            for h in range(1,25):
                date_list.append(f"20{y:0>2}-{m:0>2}-{d:0>2} {h:0>2}")

    for m in [2]:
        for d in range(1,29):
            for h in range(1,25):
                date_list.append(f"20{y:0>2}-{m:0>2}-{d:0>2} {h:0>2}")

for y in [16]:
    for m in [1,3,5,7,8,10,12]:
        for d in range(1,32):
            for h in range(1,25):
                date_list.append(f"20{y:0>2}-{m:0>2}-{d:0>2} {h:0>2}")

for m in [4,6,9,11]:
    for d in range(1,31):
        for h in range(1,25):
            date_list.append(f"20{y:0>2}-{m:0>2}-{d:0>2} {h:0>2}")

for m in [2]:
    for d in range(1,30):
        for h in range(1,25):
            date_list.append(f"20{y:0>2}-{m:0>2}-{d:0>2} {h:0>2}")

date_list.sort()
```
