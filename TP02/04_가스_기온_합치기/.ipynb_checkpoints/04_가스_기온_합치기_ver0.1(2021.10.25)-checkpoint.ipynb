```python
import pandas as pd
from numpy import nan
```

```python
total = pd.read_csv('../data/한국가스공사_시간별 공급량_20181231.csv', encoding='cp949')
total.head()
```

```python
temp = pd.read_csv('../기온자료/2013-2018년_시간별_기온2.csv', encoding='utf-8')
temp.info()
```

### 가스 공급량 라벨인코딩

```python
total['구분'].unique()
```

```python
# 구분 열을 숫자로 바꾸기
d_map = {}
for i, d in enumerate(total['구분'].unique()):
    d_map[d] = i
total['구분'] = total['구분'].map(d_map)
```

```python
total
```

```python
# 연월일 열은 object형이므로 년, 월, 일로 나눈다.
total['연월일'] = pd.to_datetime(total['연월일'])
total['year'] = total['연월일'].dt.year
total['month'] = total['연월일'].dt.month
total['day'] = total['연월일'].dt.day
total['weekday'] = total['연월일'].dt.weekday
```

```python
total
```

```python
total.info()
```

```python
temp.head()
```

```python
gas_temp = pd.merge(total, temp, on=['year', 'month', 'day', '시간'])
gas_temp
```

```python
gas_temp2 = gas_temp.copy()
```

```python
gas_temp2 = gas_temp2.drop(["Unnamed: 0", 'Unnamed: 0.1', '연-월-일 시간'], axis=1)
gas_temp2
```

### 샘플 자료형 변환

```python
gas_temp2.info()
```

```python
gas_temp2['기온'].isnull().sum()
```

```python
gas_temp3 = gas_temp2.copy()
```

```python
gas_temp3['기온'] = gas_temp3['기온'].apply(pd.to_numeric, errors='coerce')
gas_temp3.info()
```

```python
gas_temp3['기온'].isnull().sum()
```

```python
gas_temp3.to_csv('2013-2018년_가스공급량과_기온1.csv')
```
