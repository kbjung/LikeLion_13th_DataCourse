# 2013~2018년 기상청자료와 한국가스공사자료 합치기

```python
import pandas as pd
```

## 기상청 데이터

```python
raw_temp_data = pd.read_csv('../기상청기온자료/2013-2018년_기상청기온자료02.csv')
raw_temp_data.head()
```

```python
raw_temp_data.info()
```

## 필요한 특성만 선택

```python
temp_data = raw_temp_data[['일시', '기온(°C)', '습도(%)', '현지기압(hPa)']]
temp_data.head()
```

```python
temp_data.columns = ['일시', '기온', '습도', '기압']
temp_data.head()
```

```python
temp_data.info()
```

```python
temp_data[['year', 'month', 'day', '시간']] = temp_data['일시'].str.split(r'-| ').tolist()
temp_data.head()
```

## 한국가스공사 가스공급량 데이터

```python
raw_gas_data = pd.read_csv('../data/한국가스공사_시간별 공급량_20181231.csv', encoding='cp949')
raw_gas_data.head()
```

```python
raw_gas_data['구분'].unique()
```

```python
# 구분 열을 숫자로 바꾸기
d_map = {}
for i, d in enumerate(raw_gas_data['구분'].unique()):
    d_map[d] = i
raw_gas_data['구분'] = raw_gas_data['구분'].map(d_map)
raw_gas_data.head()
```

```python
# 연월일 열은 object형이므로 년, 월, 일로 나눈다.
raw_gas_data['연월일'] = pd.to_datetime(raw_gas_data['연월일'])
raw_gas_data['year'] = raw_gas_data['연월일'].dt.year
raw_gas_data['month'] = raw_gas_data['연월일'].dt.month
raw_gas_data['day'] = raw_gas_data['연월일'].dt.day
raw_gas_data['weekday'] = raw_gas_data['연월일'].dt.weekday
raw_gas_data.head()
```

## 2013-2018년 가스공급량과 기상청 데이터 합치기

```python
temp_data.info()
```

```python
temp_data[['year','month','day','시간']] = temp_data[['year','month','day','시간']].astype(int)
temp_data.info()
```

```python
raw_gas_data.info()
```

```python
gas_temp = pd.merge(raw_gas_data, temp_data, on=['year', 'month', 'day', '시간'])
gas_temp.head()
```

```python
# 필요없는 열 제거
gas_temp = gas_temp.drop(['일시'], axis=1)
gas_temp.head()
```

```python
# 열 순서 정리
gas_temp = gas_temp[['연월일', 'year','month','day','weekday','시간','구분', '공급량', '기온', '습도','기압']]
gas_temp.head()
```

```python
gas_temp.info()
```

# 이상치 처리

```python
# 공급량 이상치
# 구분0 : 4월에 10,000초과 값 존재
    # 1개
# 구분1 : 7월에 1,250초과 값 존재
    # 9개
# 구분2 : 4월 50미만, 5월 20미만, 9월 20미만, 10월 25미만, 12월 10미만
    # 226 + 900 + 38 + 6 + 46 = 1,216개
# 구분3
# 구분4
# 구분5 : 1,2,11월에 7,000초과
    # 4개
# 구분6 : 1,8월에 2,000초과, 4월에 900초과, 6월 250초과 1개, 10월 500초과 1개
    # 2 + 2 + 2 + 1 + 1 = 8개
# 전체에서 공급량 3미만 값 제거 : 1,542개
# 총 2,780개

# 남는 행 : 368,088 - 2,780 = 365,308개
```

```python
# 구분0 : 4월에 10000초과 값 존재, 1개
div0_4index = gas_temp[(gas_temp['구분']==0) & (gas_temp['month']==4)&(gas_temp['공급량'] > 10000)].index
print(len(div0_4index), div0_4index)
```

```python
gas_temp = gas_temp.drop(div0_4index)
gas_temp.info()
```

```python
# 구분1 : 7월에 1250초과 값 존재, 9개
div1_7index = gas_temp[(gas_temp['구분']==1) & (gas_temp['month']==7)&(gas_temp['공급량'] > 1250)].index
print(len(div1_7index), div1_7index)
gas_temp = gas_temp.drop(div1_7index)
gas_temp.info()
```

```python
# 구분2 : 4월 50미만, 226개
div2_4index = gas_temp[(gas_temp['구분']==2) & (gas_temp['month']==4)&(gas_temp['공급량'] < 50)].index
print(len(div2_4index), div2_4index)
gas_temp = gas_temp.drop(div2_4index)
gas_temp.info()
# 5월 20미만
# 9월 20미만
# 10월 25미만
# 12월 10미만
```

```python
# 구분2 : 4월 50미만
# 5월 20미만, 900개
div2_5index = gas_temp[(gas_temp['구분']==2) & (gas_temp['month']==5)&(gas_temp['공급량'] < 20)].index
print(len(div2_5index), div2_5index)
gas_temp = gas_temp.drop(div2_5index)
gas_temp.info()
# 9월 20미만
# 10월 25미만
# 12월 10미만
```

```python
# 구분2 : 4월 50미만
# 5월 20미만
# 9월 20미만, 38개
div2_9index = gas_temp[(gas_temp['구분']==2) & (gas_temp['month']==9)&(gas_temp['공급량'] < 20)].index
print(len(div2_9index), div2_9index)
gas_temp = gas_temp.drop(div2_9index)
gas_temp.info()
# 10월 25미만
# 12월 10미만
```

```python
# 구분2 : 4월 50미만
# 5월 20미만
# 9월 20미만
# 10월 25미만, 6개
div2_10index = gas_temp[(gas_temp['구분']==2) & (gas_temp['month']==10)&(gas_temp['공급량'] < 25)].index
print(len(div2_10index), div2_10index)
gas_temp = gas_temp.drop(div2_10index)
gas_temp.info()
# 12월 10미만
```

```python
# 구분2 : 4월 50미만
# 5월 20미만
# 9월 20미만
# 10월 25미만
# 12월 10미만, 46개
div2_12index = gas_temp[(gas_temp['구분']==2) & (gas_temp['month']==12)&(gas_temp['공급량'] < 10)].index
print(len(div2_12index), div2_12index)
gas_temp = gas_temp.drop(div2_12index)
gas_temp.info()
```

```python
# 구분5 : 1,2,11월에 7000초과, 4개
div5_index = gas_temp[(gas_temp['구분']==5) &(gas_temp['공급량'] > 7000)].index
print(len(div5_index), div5_index)
gas_temp = gas_temp.drop(div5_index)
gas_temp.info()
```

```python
# 구분6 : 1월 2000초과, 2개
div6_1index = gas_temp[(gas_temp['구분']==6) & (gas_temp['month']==1)&(gas_temp['공급량'] > 2000)].index
print(len(div6_1index), div6_1index)
gas_temp = gas_temp.drop(div6_1index)
gas_temp.info()
```

```python
# 구분6 : 4월 900초과, 2개
div6_4index = gas_temp[(gas_temp['구분']==6) & (gas_temp['month']==4)&(gas_temp['공급량'] > 900)].index
print(len(div6_4index), div6_4index)
gas_temp = gas_temp.drop(div6_4index)
gas_temp.info()
```

```python
# 구분6 : 6월 250초과 1개
div6_6index = gas_temp[(gas_temp['구분']==6) & (gas_temp['month']==6)&(gas_temp['공급량'] > 250)].index
print(len(div6_6index), div6_6index)
gas_temp = gas_temp.drop(div6_6index)
gas_temp.info()
```

```python
# 구분6 : 8월 2000초과, 2개
div6_8index = gas_temp[(gas_temp['구분']==6) & (gas_temp['month']==8)&(gas_temp['공급량'] > 2000)].index
print(len(div6_8index), div6_8index)
gas_temp = gas_temp.drop(div6_8index)
gas_temp.info()
```

```python
# 구분6 : 10월 500초과 1개
div6_10index = gas_temp[(gas_temp['구분']==6) & (gas_temp['month']==10)&(gas_temp['공급량'] > 500)].index
print(len(div6_10index), div6_10index)
gas_temp = gas_temp.drop(div6_10index)
gas_temp.info()
```

```python
# 전체에서 공급량 3미만 값 제거
indexs = gas_temp[(gas_temp['공급량'] < 3)].index
print(len(indexs), indexs)
gas_temp = gas_temp.drop(indexs)
gas_temp.info()
# 365308개
```
