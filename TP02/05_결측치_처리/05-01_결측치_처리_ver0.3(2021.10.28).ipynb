```python
# 결측치 채우기 함수화
def nan_fill(month, day, hour):
    # 해당 시간의 전후시간 평균 구하기
    # 전날
    nan_temp = total.loc[(total['month']==month) & (total['day']==day) & (total['시간']==hour)]['기온'].mean()
    
    # 해당 결측값 평균값으로 채우기
    total.loc[(total['month']==month) & (total['day']==day) & (total['시간']==hour) & (total['기온'].isnull()), '기온'] = nan_temp
    return total
```

```python
# 결측값 확인
total[total['기온'].isnull()][:10]
```

```python
# 2013-2018년
# 31일까지 있는 달
# 1-31일까지 처리
# 1-24시까지 처리
for month in [1,3,5,7,8,10,12]:
    for day in range(1,32):
        for hour in range(1,25):
            total = nan_fill(month, day, hour)
```

```python
# 확인1
total.info()
# 기온 유효 값 : 366975 -> 367409
```

```python
# 2013-2018년
# 30일까지 있는 달
# 1-30일까지 처리
# 1-24시까지 처리

for m in [4,6,9,11]:
    for d in range(1,31):
        for h in range(1,25):
            total = nan_fill(m, d, h)
```

```python
# 2013-2018년
# 2월
# 1-28일까지 처리(2016년 2월은 29일까지 있지만 결측값이 있으면 다시 처리)
# 1-24시까지 처리
for m in [2]:
    for d in range(1,29):
        for h in range(1,25):
            total = nan_fill(m, d, h)
```

```python
# 12월 31일 24시 데이터는 사이트에서 없었음.
# 해당 날짜 앞뒤 날의 기온 평균으로 결측값 채우려 했으나, 2018년 12월 31일 다음 날의 데이터가 없음.
# 따라서 전시간인 12월 31일 23시 평균 기온으로 채움.
# 연도별 12월 31일 기온 차가 나는 편. 따라서 연도별로 평균낼 것.
```

```python
# 12월 31일 24시 결측값 처리
for y in range(2013,2019):
    # 해당 월, 일, 시간의 평균 구하기
    nan_temp = total.loc[(total['year']==y) & (total['month']==12) & (total['day']==31) & (total['시간']==23)]['기온'].mean()
    # 해당 월, 일, 시간 결측값 평균값으로 채우기
    total.loc[(total['year']==y) & (total['month']==12) & (total['day']==31) & (total['시간']==24) & (total['기온'].isnull()), '기온'] = nan_temp
```

```python
# 최종확인1
total.info()
# 기온 유효 값 : 368046 -> 368088
# 결측값 모두 채워짐.
```
