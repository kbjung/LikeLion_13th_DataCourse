# 기상청기온자료 결측치 처리
* 결측치
* 기온 : 4+1 = 5
* 습도 : 4+17 = 21
* 현지기압 : 2+5 = 7

```python
import pandas as pd
```

```python
total_temp = pd.read_csv('../기상청기온자료/2013-2018년_기상청기온자료01.csv')
total_temp.head()
```

```python
total_temp.info()
```

### 기온 결측치 5개 처리

```python
total_temp[total_temp['기온(°C)'].isnull()]
```

```python
total_temp['기온(°C)'] = total_temp['기온(°C)'].interpolate()
total_temp[total_temp['기온(°C)'].isnull()]
```

```python
total_temp.info()
```

### 습도 21개 결측치 처리

```python
total_temp[total_temp['습도(%)'].isnull()]
```

```python
total_temp['습도(%)'] = total_temp['습도(%)'].interpolate()
total_temp[total_temp['습도(%)'].isnull()]
```

```python
total_temp.info()
```

### 현지기압 결측치 7개 처리

```python
total_temp[total_temp['현지기압(hPa)'].isnull()]
```

```python
total_temp['현지기압(hPa)'] = total_temp['현지기압(hPa)'].interpolate()
total_temp[total_temp['현지기압(hPa)'].isnull()]
```

```python
total_temp.info()
```

### 파일로 출력하기

```python
total_temp.to_csv('2013-2018년_기상청기온자료02.csv', index=False)
```
