```python
# 결측치 채우기 함수화
def nan_fill(month, day, time):
    # 해당 월, 일, 시간의 평균 구하기
    nan_temp = total.loc[(total['month']==month) & (total['day']==day) & (total['시간']==time)]['기온'].mean()
    # 2013년 1월 9일 14시 채우기
    total.loc[(total['month']==month) & (total['day']==day) & (total['시간']==time) & (total['기온'].isnull())] = nan_temp
    return total
```
