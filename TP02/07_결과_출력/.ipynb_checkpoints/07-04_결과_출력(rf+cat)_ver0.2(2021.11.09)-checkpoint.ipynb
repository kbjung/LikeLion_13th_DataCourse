# [RandomForest]

# 가스공급량과 기온 데이터 불러오기(2013-2018년)

```python
import pandas as pd
```

```python
df = pd.read_csv('../2013-2018년_가스공급량과_기온2.csv')
df.head()
```

# 랜덤포레스트회귀 모델 기온 예측(2013-2018년)

### 훈련(2013-2018년)

```python
from sklearn.model_selection import train_test_split
```

```python
col = ['month', '구분', '시간']
X = df[col]
y = df['기온']
```

### 정규화, 표준화

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
mms_X = MinMaxScaler().fit_transform(X)
nor_X = StandardScaler().fit_transform(mms_X)
```

```python
# nor_X
```

### 다항 특성

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pol_X = PolynomialFeatures(degree=2, include_bias=False).fit_transform(nor_X)
```

### 특성 선택(일변량 통계)

```python
from sklearn.feature_selection import SelectPercentile, f_regression
```

```python
X_train, X_test, y_train, y_test = train_test_split(pol_X, y, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
select = SelectPercentile(score_func=f_regression, percentile=50)
select.fit(X_train, y_train)
selected_X_train = select.transform(X_train)
selected_X_test = select.transform(X_test)
```

```python
selected_X_train.shape, selected_X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
```

```python
rf_r = RandomForestRegressor()
rf_r.fit(selected_X_train, y_train)
kfold = KFold(shuffle=True, random_state=0)
score1 = cross_val_score(rf_r, selected_X_train, y_train, cv=kfold).mean()
score2 = cross_val_score(rf_r, selected_X_test, y_test, cv=kfold).mean()
```

```python
print(f"훈련 셋 결정계수 = {score1}")
print(f"테스트 셋 결정계수 = {score2}")
```

# [CatBoost]

# 가스공급량과 기온 데이터 불러오기(2013-2018년)

```python
df = pd.read_csv('../2013-2018년_가스공급량과_기온2.csv')
df.head()
```

## catboost회귀 모델 기온 예측(2013-2018년)

### 훈련(2013-2018년)

```python
col = ['month', '구분', '시간']
X = df[col]
y = df['기온']
```

```python
# 정규화
mms_X = MinMaxScaler().fit_transform(X)
# 표준화
nor_X = StandardScaler().fit_transform(mms_X)
# 다항 특성
pol_X = PolynomialFeatures(degree=2, include_bias=False).fit_transform(nor_X)
# 훈련, 검증 셋 분할
X_train, X_test, y_train, y_test = train_test_split(pol_X, y, random_state=0)

# 특성 선택
select = SelectPercentile(score_func=f_regression, percentile=50)
select.fit(X_train, y_train)
selected_X_train = select.transform(X_train)
selected_X_test = select.transform(X_test)
# 확인
selected_X_train.shape, selected_X_test.shape
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold, cross_val_score
```

```python
cb = CatBoostRegressor()
cb.fit(selected_X_train, y_train)
kfold = KFold(shuffle=True, random_state=0)
score1 = cross_val_score(cb, selected_X_train, y_train, cv=kfold).mean()
score2 = cross_val_score(cb, selected_X_test, y_test, cv=kfold).mean()
```

```python
print(f"훈련 셋 결정계수 = {score1}")
print(f"테스트 셋 결정계수 = {score2}")
```

# catboost회귀 모델 가스공급량 훈련(2013-2018)

### 기온 포함해서 훈련(2013-2018)

```python
# 훈련 특성(기온 포함)
col = ['month', '시간', '구분', '기온']
X = df[col]
y = df['공급량']
```

```python
# 정규화
mms_X = MinMaxScaler().fit_transform(X)
# 표준화
nor_X = StandardScaler().fit_transform(mms_X)
# 다항 특성
pol_X = PolynomialFeatures(degree=2, include_bias=False).fit_transform(nor_X)
# 훈련, 검증 셋 분할
X_train, X_test, y_train, y_test = train_test_split(pol_X, y, random_state=0)

# 특성 선택
select = SelectPercentile(score_func=f_regression, percentile=50)
select.fit(X_train, y_train)
selected_X_train = select.transform(X_train)
selected_X_test = select.transform(X_test)
# 확인
selected_X_train.shape, selected_X_test.shape
```

```python
cb = CatBoostRegressor()
cb.fit(selected_X_train, y_train)
kfold = KFold(shuffle=True, random_state=0)
score1 = cross_val_score(cb, selected_X_train, y_train, cv=kfold).mean()
score2 = cross_val_score(cb, selected_X_test, y_test, cv=kfold).mean()
```

```python
print(f"훈련 셋 결정계수 = {score1}")
print(f"테스트 셋 결정계수 = {score2}")
```
