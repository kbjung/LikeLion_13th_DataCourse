```python
import plotly
import cufflinks as cf
import pandas as pd
```

```python
# 오프라인 모드에서도 인터렉티브한 그래픽을 가능하도록 하기
# Enabling the offline mode for interactive plotting locally
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
 # 파일명 : 온라인쇼핑몰_취급상품범위별_상품군별거래액(20.01-21.07)
mer = pd.read_csv('온라인쇼핑몰_취급상품범위별_상품군별거래액(20.01-21.07).csv', encoding='cp949')
mer.head()
```

```python
# 합계 열 빼기
mer2 = mer.drop(['합계'], axis=1)
mer2
```

```python
mer2.index
```

```python
# 계 행 없애기
mer2 = mer2.drop([0])
mer2
```

```python
# 값을 숫자형으로 인식시키기
# 테스트(결과 : 한 열씩 진행해야 함.)
mer2.iloc[:,1:].astype(int)
mer2.info()
```

```python
mer2.columns
```

```python
cols = mer2.columns
for one in cols:
    if one != '시점':
        mer2[one] = mer2[one].astype(int)

mer2.info()
```

```python
# 그래프 그리기
mer2.iplot(kind='bar', x="시점")
```

```python
col = ['컴퓨터 및 주변기기', '스포츠·레저용품', '여행 및 교통서비스', '음·식료품', '가전·전자·통신기기']
mer2.iplot(kind='line', x="시점", y=col, theme='space')
```
