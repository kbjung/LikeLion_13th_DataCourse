```python
import pandas as pd
```

```python
# 파일 불러오기
sub = pd.read_csv("CARD_SUBWAY_MONTH(20.01-21.08)(utf-8).csv")
sub.head()
```

#### 월별로 승차총승객수 분류

```python
# 사용일자 문자타입 : int64
# 2020년 01월 테스트
m2001 = sub[sub['사용일자'] <= 20200131]
m2001.head(), m2001.tail()
```

```python
# 2020년 01월부터 2021년 08월까지 월별로 자료 분류
m2001 = sub[sub['사용일자'] <= 20200131]
m2002 = sub[(sub['사용일자'] > 20200131) & (sub['사용일자'] <= 20200229)]
m2003 = sub[(sub['사용일자'] > 20200229) & (sub['사용일자'] <= 20200331)]
m2004 = sub[(sub['사용일자'] > 20200331) & (sub['사용일자'] <= 20200430)]
m2005 = sub[(sub['사용일자'] > 20200430) & (sub['사용일자'] <= 20200531)]
m2006 = sub[(sub['사용일자'] > 20200531) & (sub['사용일자'] <= 20200630)]
m2007 = sub[(sub['사용일자'] > 20200630) & (sub['사용일자'] <= 20200731)]
m2008 = sub[(sub['사용일자'] > 20200731) & (sub['사용일자'] <= 20200831)]
m2009 = sub[(sub['사용일자'] > 20200831) & (sub['사용일자'] <= 20200930)]
m2010 = sub[(sub['사용일자'] > 20200930) & (sub['사용일자'] <= 20201031)]
m2011 = sub[(sub['사용일자'] > 20201031) & (sub['사용일자'] <= 20201130)]
m2012 = sub[(sub['사용일자'] > 20201130) & (sub['사용일자'] <= 20201231)]
m2101 = sub[(sub['사용일자'] > 20201231) & (sub['사용일자'] <= 20210131)]
m2102 = sub[(sub['사용일자'] > 20210131) & (sub['사용일자'] <= 20210228)]
m2103 = sub[(sub['사용일자'] > 20210228) & (sub['사용일자'] <= 20210331)]
m2104 = sub[(sub['사용일자'] > 20210331) & (sub['사용일자'] <= 20210430)]
m2105 = sub[(sub['사용일자'] > 20210430) & (sub['사용일자'] <= 20210531)]
m2106 = sub[(sub['사용일자'] > 20210531) & (sub['사용일자'] <= 20210630)]
m2107 = sub[(sub['사용일자'] > 20210630) & (sub['사용일자'] <= 20210731)]
m2108 = sub[(sub['사용일자'] > 20210731) & (sub['사용일자'] <= 20210831)]
```

```python
# 월별 승차총승객수와 하차총승객수 각 합계를 데이터 프레임으로 만들기
# 월별 승차 인원
month_in = []
month_in.append(m2001.sum()['승차총승객수'])
month_in.append(m2002.sum()['승차총승객수'])
month_in.append(m2003.sum()['승차총승객수'])
month_in.append(m2004.sum()['승차총승객수'])
month_in.append(m2005.sum()['승차총승객수'])
month_in.append(m2006.sum()['승차총승객수'])
month_in.append(m2007.sum()['승차총승객수'])
month_in.append(m2008.sum()['승차총승객수'])
month_in.append(m2009.sum()['승차총승객수'])
month_in.append(m2010.sum()['승차총승객수'])
month_in.append(m2011.sum()['승차총승객수'])
month_in.append(m2012.sum()['승차총승객수'])
month_in.append(m2101.sum()['승차총승객수'])
month_in.append(m2102.sum()['승차총승객수'])
month_in.append(m2103.sum()['승차총승객수'])
month_in.append(m2104.sum()['승차총승객수'])
month_in.append(m2105.sum()['승차총승객수'])
month_in.append(m2106.sum()['승차총승객수'])
month_in.append(m2107.sum()['승차총승객수'])
month_in.append(m2108.sum()['승차총승객수'])

#월별 하차 인원
month_out = []
month_out.append(m2001.sum()['하차총승객수'])
month_out.append(m2002.sum()['하차총승객수'])
month_out.append(m2003.sum()['하차총승객수'])
month_out.append(m2004.sum()['하차총승객수'])
month_out.append(m2005.sum()['하차총승객수'])
month_out.append(m2006.sum()['하차총승객수'])
month_out.append(m2007.sum()['하차총승객수'])
month_out.append(m2008.sum()['하차총승객수'])
month_out.append(m2009.sum()['하차총승객수'])
month_out.append(m2010.sum()['하차총승객수'])
month_out.append(m2011.sum()['하차총승객수'])
month_out.append(m2012.sum()['하차총승객수'])
month_out.append(m2101.sum()['하차총승객수'])
month_out.append(m2102.sum()['하차총승객수'])
month_out.append(m2103.sum()['하차총승객수'])
month_out.append(m2104.sum()['하차총승객수'])
month_out.append(m2105.sum()['하차총승객수'])
month_out.append(m2106.sum()['하차총승객수'])
month_out.append(m2107.sum()['하차총승객수'])
month_out.append(m2108.sum()['하차총승객수'])

sub_month = {'월별승차총승객수': month_in, '월별하차총승객수' : month_out}
sub_month
```
