```python
from gensim import corpora
from gensim import models
```

```python
import pandas as pd
```

```python
documents3 = pd.read_csv('2020년08월_집단감염_뉴스(구글).csv')
documents3.shape
```

```python
documents3
```

```python
documents =[]
for i in range(100):
    tmp = documents3.iloc[i,0]
    documents.append(tmp)
print(type(documents), documents)
```

### 불용어 처리

```python
stoplist = ('.','!', '?', ',', ':', '[', ']', '"', '"', '...', '-', '<', '…', '–',
            '집단감염', '코로나19', '두', '집단', '최대', '혹시', '못', '또',
            '다시', '어떻게', '비상', '감염', '확진', '확진자', "'코로나19'",
            '방역', '같은', "'집단감염'", '집단감염,', '코로나19:')  # 불용어 리스트 

# 단어 리스트 만들기
texts = [
         [word for word in document.split() if word not in stoplist]  for document in documents
        ]
texts
```

```python
dictionary = corpora.Dictionary(texts)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
print(type(corpus), 'corpus : {}'.format(corpus))
```

```python
lda = models.ldamodel.LdaModel( corpus=corpus,
                                id2word=dictionary,
                                num_topics=2, # 몇 개의 주제어로 묶을 건지 설정
                                random_state = 1)
```

```python
lda.show_topics()
```

## 시각화

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'
```

```python
wc = WordCloud(background_color='white',
               max_words=200,
               contour_width=3,
               contour_color='steelblue', 
               font_path=path)
```

```python
fig = plt.figure(figsize=(30,30))

for t in range(lda.num_topics):
  plt.subplot(5,4,t+1)
  x = dict(lda.show_topic(t,200))    # 주제별 200개의 단어 보기
  im = wc.generate_from_frequencies(x)
  plt.imshow(im)
  plt.axis("off")
  plt.title("Topic #" + str(t))

plt.savefig('LDA_wordcloud.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close(fig)
```
