### 분노의 질주 댓글 파일을 가져와 시각화 하기

```python
# 댓글 파일 읽고, 담기.
doc_ko = open("15_TheExtreme_utf8.txt").read()
doc_ko[1:1000]
```

### Okt로 텍스트 파일을 담아 정리하기

```python
from konlpy.tag import Okt
t = Okt()
```

```python
doc_nouns = t.nouns(doc_ko)
print(len(doc_nouns))
doc_nouns
```

### nltk(자연어 처리 라이브러리)로 추가 정리하기

```python
# 자연어 처리 라이브러리
import nltk
```

```python
ko = nltk.Text(doc_nouns, name='분노의 질주') # nltk 객체 생성
type(ko)
```

```python
print(len(ko.tokens), type(ko.tokens))
print(ko.tokens[0:15]) # 단어를 보여줌
print(ko.vocab().most_common(20)) # 가장 많이 나온 단어와 그 횟수를 보여줌
```

```python
most_fre = ko.vocab().most_common(50)
most_fre
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stopwords = ['이', '것', '점', '영화', '볼']

new_ko = []
for one_word in ko:
    if one_word not in stopwords: # 불용어 단어 리스트에 있는지 확인
        new_ko.append(one_word)
```

```python
len(new_ko)
```

```python
new_ko = nltk.Text(new_ko, name='분노의 질주2')
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```python
plt.figure(figsize=(15,8))
new_ko.dispersion_plot(['자동차', '워커', '재미', '평점'])
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
Car_mask = np.array(Image.open("Draw_car1.png"))
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
wc = WordCloud(max_words=500,
               mask=Car_mask,
               font_path=path,
               background_color='white',
               contour_width=3,
               contour_color='steelblue').generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis('off')
plt.show
```
