```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# np.linsapce(시작, 끝, 데이터 개수)
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
```

```python
x
```

```python
plt.plot(x, y1)
```

##### 1-2 실습 - 3개 그래프 그리기

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4
plt.plot(x, y1, x, y2, x, y3)
```

##### 1-2 (추가) cos, tan 그래프 그려보기

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.plot(x, y1, x, y2, x, y3)
```

###### 1-2 (추가) 선그래프가 아닌 점 형태로 plot해보기

```python
# plt.plot(x축 값, y축 값, '그래프 형태')
plt.plot(x, y1, 'o')
plt.plot(x, y2, 'o')
plt.plot(x, y3, 'o')
```
