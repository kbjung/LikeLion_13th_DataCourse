```python
import seaborn as sns
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

t = sns.load_dataset('titanic')
```

```python
# 클래스 별 요금의 평균, 최고, 최저 가
fare_by_class = t[["fare","age", "sex", "pclass"]].groupby(["pclass"])
fare_by_class
```

```python
# 클래스 별 평균
fare_by_class["fare"].mean()
```

```python
# 클래스 별 최댓값
fare_by_class["fare"].max()
```

```python
# 클래스 별 최솟값
fare_by_class["fare"].min()
```

```python
# 등석 별 성별, 나이, 요금
sel = ["sex", "age", "fare"]
fare_by_sex_age1 = t[sel].loc[t["pclass"]==1,:]
fare_by_sex_age2 = t[sel].loc[t["pclass"]==2,:]
fare_by_sex_age3 = t[sel].loc[t["pclass"]==3,:]

fare_by_sex_age1.shape, fare_by_sex_age2.shape, fare_by_sex_age3.shape
```

```python
# 히스토그램(x=나이, y=요금, 비교군=성별)
sns.set_style("whitegrid")
plt.figure(figsize=(12,12))

sns.lmplot(x="age", y="fare", hue="sex", data=fare_by_sex_age1)
plt.title("Class1")

sns.lmplot(x="age", y="fare", hue="sex", data=fare_by_sex_age2)
plt.title("Class2")

sns.lmplot(x="age", y="fare", hue="sex", data=fare_by_sex_age3)
plt.title("Class3")
```
