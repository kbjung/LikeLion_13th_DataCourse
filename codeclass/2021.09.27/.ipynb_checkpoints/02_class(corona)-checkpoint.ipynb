### 2-2 백신 접종률과 중증 감소의 관계는 어떠한가?

```python
corona.head(3)
```

```python
corona['위중증'].unique()
```

#### 문자열 처리

```python
corona['위중증'].str.split('\n')
```

```python
corona['위중증'].str.split('\n', expand=True) # expand : 나누어진 문자열 중 뒤의 문자열을 새로운 열을 만들어 추가. 기본값은 False
```

```python
corona[['위중증_수', '위중증_증감']] = corona['위중증'].str.split('\n', n=2, expand=True)
corona.head()
```

```python
### 위중증률 = 위중증_수/인구수
corona['위중증률'] = corona['위중증_수'] /  corona['인구수']
corona.head() # 데이터 타입이 문자열이라 문제 발생
```

```python
corona.info()
```

#### 데이터 타입 변경

```python
corona['위중증_수'] = corona['위중증_수'].astype('int64') # 비어있는 값이 있어 오류 발생
```

```python
corona.isnull().sum()
```

```python
corona[corona['위중증_수'].isnull()]
```

```python
corona.loc[corona['위중증_수'].isnull(), :] # corona[corona['위중증_수'].isnull()]와 동일한 의미
```

### 위증증_수2를 만들고, 위증증수가 없는 것은 -99999로 만들자.

```python
# 위중증_수와 같은 값을 가져오면서 별개로 움직이도록 설정 
# corona['위중증_수2'] = corona['위중증_수'] 로하면 '위중증_수2' 값을 변경하면 '위중증_수'도 바뀐다.
# 따라서 .copy()를 수행해야한다.
corona['위중증_수2'] = corona['위중증_수'].copy()
```

```python
corona.loc[corona['위중증_수2'].isnull(), '위중증_수2'] = -99999 #변경

# 확인
corona.loc[corona['위중증_수2'].isnull(), :]
```

```python
corona['위중증_수2'] = corona['위중증_수2'].astype('int16')
corona['위중증률'] = corona['위중증_수2'] / corona['인구수']
corona
```

```python
corona_sel = ['국가명', '발생률', '인구수', '확진자_합계', '백신접종', '인구당확진자비율',  
              '2차접종', '사망률', '위중증', '위중증_증감', '위중증_수', '위중증_수2', '위중증률']
corona_part = corona[corona_sel]
corona_part
```

```python
sns.lmplot(x='위중증률', y='2차접종', data=corona_part)
```

```python
corona_part[['위중증률', '2차접종']].corr()
```
