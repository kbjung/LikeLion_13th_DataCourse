### plotly express 이용

```python
# 라이브러리 로드
import plotly.express as px
```

```python
titanic.head()
```

```python
# 그래프로 표현하기
fig = px.scatter(titanic, x='pclass', y='fare', trendline='ols', marginal_y='violin')
fig.show()
```

```python
fig = px.scatter_matrix(titanic, dimensions=["pclass", "fare", "sex"], color='pclass')
fig.show()
```

```python
# 범주형 데이터 시각화
fig = px.parallel_categories(titanic[['embark_town', 'pclass', 'sex', 'alive']], color='pclass')
fig.show()
```
