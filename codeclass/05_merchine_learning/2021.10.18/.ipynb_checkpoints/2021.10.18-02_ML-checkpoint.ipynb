```python
import seaborn as sns
plt.figure(figsize=(12,10))
sns.heatmap(train.corr(), annot=True, fmt='.2f')
```

```python
sel = ['condition', 'floors', 'lat', 'long']
X = train[sel]
y = train['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
### 3개 모델
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print(knn.score(X_test, y_test)) # 결정계수
```

```python
mse_val = (( pred - y_test) ** 2).sum() / len(pred)
mae_val = np.abs( pred - y_test).sum() / len(pred)
print(mse_val)
print(mse_val ** 0.5)
print(mae_val)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
pred = lr_model.predict(X_test)

print(lr_model.score(X_test, y_test))
```

```python
mse_val = (( pred - y_test) ** 2).sum() / len(pred)
mae_val = np.abs( pred - y_test).sum() / len(pred)
print(mse_val)
print(mse_val ** 0.5)
print(mae_val)
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
pred = rf_model.predict(X_test)

print(rf_model.score(X_test, y_test))
```

```python
mse_val = (( pred - y_test) ** 2).sum() / len(pred)
mae_val = np.abs( pred - y_test).sum() / len(pred)
print(mse_val)
print(mse_val ** 0.5)
print(mae_val)
```

```python
print(rf_model.feature_importances_)
print(sel)
```

## RandomForest

```python
%%time # 작동 시간 측정

rf_model = RandomForestRegressor(random_state=0, n_estimators=500)
rf_model.fit(X_train, y_train)
pred = rf_model.predict(X_test)

print(rf_model.score(X_test, y_test))
```

```python
%%time

rf_model = RandomForestRegressor(random_state=0, n_estimators=500, n_jobs=-1)
rf_model.fit(X_train, y_train)
pred = rf_model.predict(X_test)

print(rf_model.score(X_test, y_test))
```
