```python
tesla_text = open("tesla_news_utf8.txt").read()
tesla_text[1:100]
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이미지를 열어 담는다.
model_X_img = Image.open('tesla_model_X_mod.png')
type(model_X_img), model_X_img
```

```python
# 이미지를 코드로 보여준다.
model_X_mask = np.array(model_X_img)
model_X_mask
```

```python
plt.figure(figsize=(15,8))
# interpolation='bilinear' 이미지 손상이 있다면, bilinear에 따라 보안해서 표현
plt.imshow(model_X_mask, interpolation='bilinear')
# 축 표시 없애기
plt.axis('off')
plt.show()
```

```python
# contour : 외각선
wc = WordCloud(max_words=2000,
               mask=model_X_mask,
               background_color='white',
               contour_width=3,
               contour_color='steelblue') # 객체 생성
wc.generate(tesla_text) # 생성된 객체를 이용해서 데이터를 넣기.
print(type(wc.words_))
print(wc.words_)
```

```python
plt.figure(figsize=(15,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
