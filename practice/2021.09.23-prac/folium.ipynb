```python
# folium 로드
import folium
```

```python
# 지도 처음 화면 설정
m = folium.Map([37.546817, 126.911411], zoom_start=10)
# 동일한 결과
# m = folium.Map(location = [37.546817, 126.911411], zoom_start=10)
m
```

```python
# 마크 찍기
# 롯데 월드 타워 : 37.512706500791, 127.10257850098388
folium.Marker([37.512706500791, 127.10257850098388], icon = folium.Icon(color="orange")).add_to(m)
m
```

```python
# tooltip 표시(마우스 올리면 "텍스트" 표시)
folium.Marker([37.512706500791, 127.10257850098388], tooltip="롯데 월드 타워", icon = folium.Icon(color="orange")).add_to(m)
m
```

```python
# popup 표시(클릭하면 "텍스트" 표시)
# 양재 KFC : 37.48433236493403, 127.03517720809431
folium.Marker([37.48433236493403, 127.03517720809431], popup="KFC", icon=folium.Icon(color="blue")).add_to(m)
m
```

```python
# 맵 "파일명.html" 파일로 저장하기
m.save("prac.html")
```
